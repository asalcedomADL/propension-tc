# Processing clientes

```python
########## Cargando paquetes ##########
import pandas as pd
import numpy as np
import pickle
import datetime
import os
import gc
```

```python
## Garbage Collector
gc.collect()
```

```python
# reading the objects containing the target
path = '/mnt/s3-refined-popular/propension/raw_piloto/cliente/'
os.chdir(path) #TODO: adjust this line #TODO: adjust this line
```

```python
# Year and Month:
year = '2019'
month = '07'
```

## Base Personas:

```python
file_path = os.path.join(path,'persona',year,month,'data.csv')
data_person = pd.read_csv(file_path, sep = '|', encoding = 'utf_8')
data_person.columns = data_person.columns.str.lower()
```

```python
## Modificando base:

data_person = data_person.rename(columns={'cont_id':'id_cliente'})
data_person = data_person.set_index('id_cliente')
data_person['orden'] = data_person.groupby('id_cliente').cumcount()
data_person = data_person[data_person['orden'] == 0]
data_person = data_person.drop('orden', axis = 1)
```

```python
## Filtrando por persona Natural
data_person = data_person[data_person['client_st_tp_ds'] == 'A-Cliente Activo']
data_person = data_person[data_person['id_tp_cd'].isin([1000003,1000005])]
```

```python
data_person = data_person[["client_st_tp_ds", "since_dt", "marital_st_tp_ds", "gender_tp_code", "birth_dt", "children_ct", "birth_state_tp_ds",
                           "profession_tp_ds", "occupation_tp_ds", "housing_tp_ds", "ciiu_tp_ds",]]
```

```python
# Cambiando tipo de dato

data_person['since_dt'] = pd.to_datetime(data_person['since_dt'], dayfirst=True, errors = 'coerce')
data_person['birth_dt'] = pd.to_datetime(data_person['birth_dt'], dayfirst=True, errors = 'coerce')

data_person = data_person.drop_duplicates()
```

## Guardando Personas:

```python
out_path = os.path.join(path,'persona',year,month,'data.pickle')
pickle.dump(data_person,open(out_path, "wb" ))
```

## Base Contact Method:

```python
# Leyendo base
file_path = os.path.join(path,'metodo-contacto',year,month,'data.csv')
data_cmethod = pd.read_csv(file_path, encoding = 'utf-8', sep = '|')
data_cmethod.columns = data_cmethod.columns.str.lower()
data_cmethod.drop_duplicates(inplace=True)
```

```python
## Modificando base:

data_cmethod = data_cmethod.rename(columns={'cont_id':'id_cliente', 'preferred_ind':'prefered_contact_ind', 'country_code':'country_code_method', 'area_code':'area_code_method'}) 
data_cmethod['id_cliente'] = data_cmethod['id_cliente'].astype(np.int64)
```

```python
# NÃºmero de contactos registrados:
data_cmethod = data_cmethod.groupby("id_cliente").size()\
             .reset_index().rename(columns={0: 'num_tel_regs'})\
             .set_index('id_cliente')
```

```python
## Guardando base cmethod
out_path = os.path.join(path,'metodo-contacto',year,month,'data.pickle')
pickle.dump(data_cmethod,open(out_path, "wb" ))
```

## Base Segmento:

```python
# Leyendo base
file_path = os.path.join(path,'segmento',year,month,'data.csv')
data_segmet = pd.read_csv(file_path, encoding = 'utf-8', sep = '|')
data_segmet.drop_duplicates(inplace=True)
data_segmet = data_segmet[["CONT_ID", "SUB_SEG_TP_DS"]]
data_segmet['CONT_ID'] = data_segmet['CONT_ID'].astype(np.int64)
data_segmet.set_index("CONT_ID", inplace=True)
```

```python
## Modificando base:

data_segmet.index.names = ["id_cliente"]
data_segmet.rename(columns=lambda x: x.lower(), inplace=True)

```

```python
# Guardando base cmethod
out_path = os.path.join(path,'segmento',year,month,'data.pickle')
pickle.dump(data_segmet, open(out_path, "wb" ))
```

# Base Address:

```python
# Leyendo base
file_path = os.path.join(path,'ubicacion',year,month,'data.csv')
data_address = pd.read_csv(file_path, encoding = 'utf-8', sep = '|')
data_address["CONT_ID"] = data_address["CONT_ID"].astype(np.int64)
data_address.drop_duplicates(inplace=True)
```

```python
# Filtrando data
data_address = data_address.groupby("CONT_ID")["ADDRESS_ID"].size()\
    .reset_index()\
    .rename(
    columns={'ADDRESS_ID': 'NUM_REG'})\
    .set_index('CONT_ID')

data_address.index.names = ["id_cliente"]
data_address.rename(columns=lambda x: x.lower(), inplace=True)
```

```python
# Guardando base data_address
out_path = os.path.join(path,'ubicacion',year,month,'data.pickle')
pickle.dump(data_address, open( out_path, "wb" ))
```

## Base  Creditrisk

```python
# Leyendo base
file_path = os.path.join(path,'riesgo',year,month,'data.csv')
data_creditrisk = pd.read_csv(file_path, encoding = 'utf-8', sep = '|')
data_creditrisk.drop_duplicates(inplace=True)
```

```python
# Modificando base
data_creditrisk = data_creditrisk[["CONT_ID", "CUST_RATING"]]
# Changing data types
data_creditrisk["CONT_ID"] = data_creditrisk["CONT_ID"].astype(np.int64)
data_creditrisk["CUST_RATING"] = data_creditrisk["CUST_RATING"].astype(object)

data_creditrisk = data_creditrisk.set_index("CONT_ID")

data_creditrisk.index.names = ["id_cliente"]
data_creditrisk.rename(columns=lambda x: x.lower(), inplace=True)
```

```python
# Guardando base data_creditrisk
out_path = os.path.join(path,'riesgo',year,month,'data.pickle')
pickle.dump(data_creditrisk, open( out_path, "wb" ))
```

## Base Financials

```python
# Leyendo base
file_path = os.path.join(path,'financiera',year,month,'data.csv')
data_financials = pd.read_csv(file_path, encoding = 'utf-8', sep = '|')
data_financials = data_financials.drop_duplicates()
```

```python
# Cambiando tipo de dato
data_financials["CONT_ID"] = data_financials["CONT_ID"].astype(np.int64)
data_financials["TOTAL_INCOME"] = data_financials["TOTAL_INCOME"].astype(np.float64)
data_financials["TOTAL_EXPENSES"] = data_financials["TOTAL_EXPENSES"].astype(np.float64)
data_financials["TOTAL_ASSETS"] = data_financials["TOTAL_ASSETS"].astype(np.float64)
data_financials["TOTAL_LIABILITIES"] = data_financials["TOTAL_LIABILITIES"].astype(np.float64)
```

```python
# Filtrando columnas
data_financials = data_financials[["CONT_ID","TOTAL_INCOME","TOTAL_EXPENSES","TOTAL_ASSETS","TOTAL_LIABILITIES"]]
```

```python
data_financials = data_financials.set_index("CONT_ID")
data_financials.index.names = ["id_cliente"]
data_financials.rename(columns=lambda x: x.lower(), inplace=True)
```

```python
# Guardando base data_financials
out_path = os.path.join(path,'financiera',year,month,'data.pickle')
pickle.dump(data_financials, open( out_path, "wb" ))
```

## Uniendo Bases

```python
# Cargando bases
persona = os.path.join(path,'persona',year,month,'data.pickle')
cmethod = os.path.join(path,'metodo-contacto',year,month,'data.pickle')
segmento = os.path.join(path,'segmento',year,month,'data.pickle')
ubicacion = os.path.join(path,'ubicacion',year,month,'data.pickle')
riesgo = os.path.join(path,'riesgo',year,month,'data.pickle')
financiera = os.path.join(path,'financiera',year,month,'data.pickle')

data_person = pickle.load(open(persona, "rb" ))
data_cmethod = pickle.load(open(cmethod, "rb" ))
data_segmet = pickle.load(open(segmento, "rb" ))
data_address = pickle.load(open(ubicacion, "rb" ))
data_creditrisk = pickle.load(open(riesgo, "rb" ))
data_financials = pickle.load(open(financiera, "rb" ))
```

```python
# Join total:

data_final = data_person.join(data_cmethod, how = 'left').join(data_segmet, how = 'left').join(data_address, how = 'left').join(data_creditrisk, how = 'left').join(data_financials, how = 'left')
```

```python
import datetime
# Creando edad:

today = datetime.datetime.now()

## Edad
data_final["edad"] = (today - data_final["birth_dt"]) / np.timedelta64(365, 'D')  
## antiguedad
data_final["antiguedad_banco_anos"] = (today - data_final["since_dt"]) / np.timedelta64(365, 'D')
## Agregando region:
data_region = pickle.load(open('/mnt/work/dict_col_regions.p', 'rb'))
data_final['region_col'] = data_final['birth_state_tp_ds'].map(data_region)
```

```python
# Cambiando nombres

n_nombres = {
        "client_st_tp_ds": "estado_cliente", "marital_st_tp_ds": "estado_civil", "gender_tp_code": "genero", "children_ct": "num_hijos",
        "birth_state_tp_ds": "departamento", "profession_tp_ds": "profesion", "occupation_tp_ds": "tipo_ocupacion", "housing_tp_ds": "tipo_hogar",
        "ciiu_tp_ds": "actividad_principal", "num_tel_regs": "num_tel_regis", "sub_seg_tp_ds": "tipo_subsegmento", "num_reg": "numero_direcciones",
        "total_income": "ingresos_totales",  "total_expenses": "egresos_totales", "total_assets": "total_activos", "total_liabilities": "total_pasivos",
        "cust_rating": "calificacion_cliente"}

data_final = data_final.rename(columns=n_nombres)
```

```python
data_final = data_final[["estado_civil", "genero", "num_hijos", "departamento", "profesion", "tipo_ocupacion", "tipo_hogar",
                         "actividad_principal", "num_tel_regis", "tipo_subsegmento", "numero_direcciones", "ingresos_totales",
                         "egresos_totales", "total_activos", "total_pasivos", "calificacion_cliente", "edad", "antiguedad_banco_anos",
                         "region_col"]]
```

```python
# Guardando base cmethod
file_path = '/mnt/work/modelVersion1/objects/Cliente/clientes_consolidado_'+year+month+'.pickle'
pickle.dump(data_final, open(file_path, "wb" ))
```
