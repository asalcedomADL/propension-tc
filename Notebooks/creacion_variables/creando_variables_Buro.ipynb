```python
#Función que calcula la diferencia porcentual vs el mes anterior del buro_201806_201908[var+'lag1']
```

```python
def dif_porcentual(var):
    buro_201806_201908[var+'lag1']=np.where(buro_201806_201908['llave_lag']==1,buro_201806_201908[var].shift(1),np.nan)
    buro_201806_201908[var+'dif']=np.where( ((np.isnan(buro_201806_201908[var])) | (buro_201806_201908[var]==0)) & ((np.isnan(buro_201806_201908[var+'lag1'])) | (buro_201806_201908[var+'lag1']==0)) ,0,
                              np.where( ((np.isnan(buro_201806_201908[var])) | (buro_201806_201908[var]==0)) & ((buro_201806_201908[var+'lag1']>0)) ,-1,
                              np.where( ((np.isnan(buro_201806_201908[var+'lag1'])) | (buro_201806_201908[var+'lag1']==0)) & ((buro_201806_201908[var]>0)) ,1,
                              np.where( (buro_201806_201908[var+'lag1']>0) & (buro_201806_201908[var]>0) , (buro_201806_201908[var]-buro_201806_201908[var+'lag1'])/buro_201806_201908[var+'lag1'],99))))
    del buro_201806_201908[var+'lag1']
    
```
