```python
import pandas as pd
import s3fs
import pyarrow.parquet as pq
```

```python
### Lectura Parquet
def read_parquet(path: str, **kwargs):
    fs = s3fs.S3FileSystem()
    dataset = pq.ParquetDataset(path, filesystem=fs)
    df = dataset.read_pandas(**kwargs).to_pandas()
    return df
```
