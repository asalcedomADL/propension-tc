```python
import pandas as pd 
import numpy as np 
import datetime
from dateutil import relativedelta
import s3fs
import pyarrow.parquet as pq
fs = s3fs.S3FileSystem()

pd.set_option('max_columns',None)
pd.set_option('max_rows',100)
```

## Configurando bases TX

```python
# Leyendo base
path = "s3://adl-refined-dev-popular/parquet/TC_adquisicion/tx_pasivo_201901_202002_vars_ahr"

input = path
dataset = pq.ParquetDataset(input, filesystem=fs)
table = dataset.read()
data = table.to_pandas()

del input, dataset, table

#data = pd.read_parquet( , engine='pyarrow')
```

```python
print(data.shape)
data.head()
```

```python
data = data[data['id_cliente'].notnull()]
data['id_cliente'] = data['id_cliente'].astype('int')
data['fecha_tx_pav'] = data['PERIODO'].astype('int')
```
