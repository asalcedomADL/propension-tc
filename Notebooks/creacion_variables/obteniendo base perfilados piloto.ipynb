## Obteniendo base Piloto perfilados

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns',None)
```

```python
path = 's3://adl-refined-dev-popular/data_orig/tmp/tmpg_perfilados_M20200817.csv'

data = pd.read_csv(path,sep='|', encoding='latin-1', dtype={'NUMERO_DOC':str,'ID_CLIENTE':str,'ID_NUM_IDENTIFICACION':str,})

print(data.shape)
data
```

```python
# PORCENTAJE ESPERADO
p = 15000/data.shape[0]
print(p)
```

```python
## SELECCIONANDO 15 MIL CLIENTES
np.random.seed(seed=4321) # Semilla

#Se generan numeros aleatorios entre 0 y 1
data['ran']=np.random.random(size=data.shape[0])

#Se seleccion train : 50% 
data_muestra=data[data['ran']<=p]

data_muestra.drop(columns=['ran'],inplace=True)
print(data_muestra.shape)
```

```python
## Guardando base muestra

path = 's3://adl-refined-dev-popular/data_orig/tmp/tmp_perfilados_mustraTU_M20200817.csv'
data_muestra[['TIPO_DOC','NUMERO_DOC','ID_CLIENTE','ID_NUM_IDENTIFICACION']].to_csv(path, sep = '|', encoding= 'utf_8', index=False)
```
