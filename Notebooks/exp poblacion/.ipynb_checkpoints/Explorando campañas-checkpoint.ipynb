## Explorando bases de campañas TC

Código para explorar la información de activaciones de TC tradicional

```python
import pandas as pd
import numpy as np
import pickle
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
import pandas_profiling
import os
from datetime import datetime, timedelta
import dateutil.relativedelta
import gc

pd.set_option('display.max_columns',None)
```

```python
path_file1='s3://adl-refined-dev-popular/campanas/tc-nuevas-proc-crm/campanas-tc-nuevas-proc-crm_H20190801-20200201.csv'

data1 = pd.read_csv(path_file1, sep = '|', encoding='ISO8859-1',quoting=1, dtype={'ID_CLIENTE':str})
print(data1.shape)
data1.head()
```

```python
## tipo campaña
data1['TIPO CAMPAÑA'] = 'Preaprobados'
```

```python
path_file2='s3://adl-refined-dev-popular/campanas/tc-nuevas-proc-vend/campanas-tc-nuevas-proc-vendors_H20190101-20200301.csv'

data2 = pd.read_csv(path_file2, sep = '|', encoding='ISO8859-1',quoting=1, dtype={'ID_CLIENTE':str})
print(data2.shape)
data2.head()
```

```python
data2[data2['ID_CLIENTE']=='561352303938404302']
```

```python
(data2.loc[data2['FECHA']=='01/11/2019','ID_CLIENTE'].value_counts() == 2).sum()
```

```python
data2['TIPO CAMPAÑA'].value_counts()
```

```python
data1n = data1.drop_duplicates()
data2n = data2.drop_duplicates()

print(data1.shape)
print(data1n.shape)
print(data2.shape)
print(data2n.shape)
```

```python
data2n
```

```python
data = pd.concat([data1n[['ID_CLIENTE','FECHA','TIPO CAMPAÑA']],data2n[['ID_CLIENTE','FECHA','TIPO CAMPAÑA']]],ignore_index=True)
print(data.shape)
data.head()
```

```python
# Casting variables
data['FECHA'] = pd.to_datetime(data['FECHA'],format='%d/%m/%Y')
data['mes_campaña'] = pd.to_datetime(data['FECHA'].dt.strftime('%Y-%m'),format='%Y-%m')
```

```python
print(data.shape)
data.head()
```

```python
# Volumenes mensuales:
pd.crosstab(data['mes_campaña'],data['TIPO CAMPAÑA'])
```

```python
plt.figure(figsize=(10, 6))
data.groupby('mes_campaña').size().plot()
plt.title('Número de clientes incluídos en campañas de TC')
plt.ylabel('Número de clientes campaña')
plt.ylabel('Mes de activación')
```

```python
## Saving campañas total
data.to_parquet('/mnt/work/CU_adquisicionTC/objects/campanas_tot_2019_20.parquet',engine='pyarrow')
```
