### Filtro inicial

```python
## Leyendo base libranzas ciclo de vida

mes = '07'
anio = '2020'

path = 's3://adl-refined-dev-popular/data_parquet/productos/libranzas/productos_libranzas_dwh_M'+anio+mes
```

```python
# Leyendo base
dataset = pq.ParquetDataset(path, filesystem=fs)
table = dataset.read()
lib = table.to_pandas()
lib = lib.drop_duplicates()

del dataset, table
```

```python
print(lib.shape)
lib.head()
```

```python
lib2 = lib.copy()
lib2 = lib2[lib2['id_cliente'].notnull()]
lib2['id_cliente'] = lib2['id_cliente'].astype('int')
```

```python
ids = lib2.loc[~lib2['ds_estado_actual'].isin(['Cancelada','Castigado']),'id_cliente'].unique()

ids_final = pd.DataFrame({'id_cliente':ids})
ids_final.shape
```

```python
np.random.seed(seed=1234) # Semilla
## Muestra del 50%
#Se generan numeros aleatorios entre 0 y 1
ids_final['ran']=np.random.random(size=ids_final.shape[0])

#Se seleccion train : 50% 
ids_final['grupo'] = np.where(ids_final['ran']<=0.5,1,2)
ids_final['grupo_piloto'] = np.where(ids_final['ran']<=0.5,'control','priorizado')

ids_final.drop(columns='ran', inplace=True)
```

```python
ids_final.head(15)
```

```python
ids_final['grupo_piloto'].value_counts()
```

```python
ids_final['grupo'].value_counts()
```

```python
## Guardando bases
ids_final.to_parquet("s3://adl-refined-dev-popular/parquet/TC_adquisicion/base_pilotoAgosto",engine='pyarrow', index=False)
```
