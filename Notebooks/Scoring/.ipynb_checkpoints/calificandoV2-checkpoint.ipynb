```python
path = "s3://adl-refined-dev-popular/parquet/TC_adquisicion/score_preaprobados_202008_13500k"


data_pre_aprob = pd.read_parquet(path, engine='pyarrow')
print(data_pre_aprob.shape)
data_pre_aprob.head()
```

```python
# Grupos Probabilidad
data_pre_aprob['segmento'] = 'seg_preaprobado'

data_pre_aprob['grupo_prob'] = np.where(data_pre_aprob['decil'] >= 3, 'grupo_prob_alta',
                                        np.where(data_pre_aprob['decil'] < 3, 'grupo_prob_baja',
                                        np.nan))
```

```python
np.random.seed(seed=4321) # Semilla
# Se generan numeros aleatorios entre 0 y 1
data_pre_aprob['ran2'] = np.random.random(size=data_pre_aprob.shape[0])

data_pre_aprob['grupo_prob'] = np.where((data_pre_aprob['score'].isnull()) & (data_pre_aprob['ran2'] <= 0.3415), 'grupo_prob_alta',               
                                                                    np.where((data_pre_aprob['score'].isnull()) & (data_pre_aprob['ran2'] > 0.3415), 'grupo_prob_baja',
                                                                             data_pre_aprob['grupo_prob']))
```

```python
p1 = 11000/data_pre_aprob.shape[0]

print(p1)
```

```python
np.random.seed(seed=4321) # Semilla
# Muestra del 50%
# Se generan numeros aleatorios entre 0 y 1
data_pre_aprob['ran'] = np.random.random(size=data_pre_aprob.shape[0])
# Grupo call
data_pre_aprob['grupos'] = np.where(data_pre_aprob['ran'] <= p1, 'grupo_call', 'grupo_mkd')
```

```python
pd.crosstab(data_pre_aprob['grupos'],data_pre_aprob['grupo_prob'])
```

```python
np.random.seed(seed=4321) # Semilla
## Separando base
data_pre_aprob.columns = data_pre_aprob.columns.str.upper()

data_pre_aprob = data_pre_aprob.rename(columns={'TIPO DOC':'TIPO_IDENTIFICACION', 'NUMERO DOC':'NUMERO_IDENTIFICACION'})

data_pre_call = data_pre_aprob[data_pre_aprob['GRUPOS'] == 'grupo_call']
data_pre_call['RAN3'] = np.random.random(size=data_pre_call.shape[0])
data_pre_call['GRUPO_TRATAMIENTO'] = np.where(data_pre_call['RAN3'] <= 0.5, 'grupo_control', 'grupo_tratamiento')

data_pre_mkd = data_pre_aprob[data_pre_aprob['GRUPOS'] == 'grupo_mkd']
```

```python
## Separando en 5 bases

cols = ['TIPO_IDENTIFICACION','NUMERO_IDENTIFICACION', 'GRUPO_PROB', 'SEGMENTO', 'SCORE', 'DECIL']

## Base Mkd
base_mkd = pd.concat([data_pre_mkd[cols],data_mkd[cols]], ignore_index=True)

## Base Call-Preaprobados control
base_pre_contr = data_pre_call.loc[data_pre_call['GRUPO_TRATAMIENTO'] == 'grupo_control',cols]
## Base Call-Perfilados control
base_perf_contr = data_call.loc[data_call['GRUPO_TRATAMIENTO'] == 'grupo_control',cols]

## Base Call-Alta prob trat
base_alta_prob_preb = data_pre_call.loc[(data_pre_call['GRUPO_TRATAMIENTO'] == 'grupo_tratamiento') &
                                   (data_pre_call['GRUPO_PROB'] == 'grupo_prob_alta'),cols]
base_alta_prob_perf = data_call.loc[(data_call['GRUPO_TRATAMIENTO'] == 'grupo_tratamiento') &
                                   (data_call['GRUPO_PROB'] == 'grupo_prob_alta'),cols]

base_alta_prob = pd.concat([base_alta_prob_preb,base_alta_prob_perf], ignore_index=True)

## Base Call-Baja prob trat
base_baja_prob_preb = data_pre_call.loc[(data_pre_call['GRUPO_TRATAMIENTO'] == 'grupo_tratamiento') &
                                   (data_pre_call['GRUPO_PROB'] == 'grupo_prob_baja'),cols]
base_baja_prob_perf = data_call.loc[(data_call['GRUPO_TRATAMIENTO'] == 'grupo_tratamiento') &
                                   (data_call['GRUPO_PROB'] == 'grupo_prob_baja'),cols]

base_baja_prob = pd.concat([base_baja_prob_preb,base_baja_prob_perf], ignore_index=True)
```

```python
import s3fs
import pyarrow.parquet as pq

def spark_read_parquet(s3_url: str, **args):
    fs = s3fs.S3FileSystem()
    # Leyendo base
    dataset = pq.ParquetDataset(s3_url, filesystem=fs)
    table = dataset.read()
    dataframe = table.to_pandas()

    del dataset, table

    return dataframe
```
